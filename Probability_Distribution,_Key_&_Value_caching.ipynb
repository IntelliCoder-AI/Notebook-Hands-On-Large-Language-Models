```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
# Load model and tokenizer
tokenizer = AutoTokenizer.from_pretrained("microsoft/phi-3-mini-4k-instruct")

model = AutoModelForCausalLM.from_pretrained(
    "microsoft/phi-3-mini-4k-instruct",
    device_map="cuda",
    torch_dtype="auto",
    trust_remote_code=True
)

# Create a pipeline

generator = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    max_new_tokens=100,
    do_sample=False,
)
```

##***Autoregressive, token-by-token generation***



```python
prompt = "Write an email apologizing to Sarah for the tragic gardening mishap. Explain how it happened."
output = generator(prompt)
print(output[0]['generated_text'])
```

##***Choosing a Single Token from the Probability Distribution by Language modeling head (LM head) as next output token***




```python
prompt = "The capital of France is"
input_ids = tokenizer(prompt, return_tensors="pt").input_ids

input_ids = input_ids.to("cuda")
model_output = model.model(input_ids)
lm_head_output = model.lm_head(model_output[0])
```

```python
# Choosing the highest scoring token as next output token
token_id = lm_head_output[0,-1].argmax(-1)
tokenizer.decode(token_id)
```

```python
print(lm_head_output[0,-1])
# The printed vector has one score for every token in the model's vocabulary.
```

##***Speeding Up Generation by Caching Keys and Values***


```python
prompt = "Write a very long email apologizing to Sarah for the tragic gardening mishap. Explain how it happened."
input_ids = tokenizer(prompt, return_tensors="pt").input_ids
input_ids = input_ids.to("cuda")
```

```python
# WITH CACHING, The model only takes approx 5 sec to generate the 100 tokens.

%%timeit -n 1  # time how long the execution takes (it runs the command several times and gets the average)
generation_output = model.generate(
input_ids=input_ids,
max_new_tokens=100,
use_cache=True
)
```

```python
# WITHOUT CACHING, The model takes approx 33 sec to generate the 100 tokens

%%timeit -n 1
generation_output = model.generate(
input_ids=input_ids,
max_new_tokens=100,
use_cache=False
)
```
